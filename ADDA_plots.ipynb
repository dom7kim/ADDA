```python
import torch
import matplotlib.pyplot as plt

def plot_learning_curve(filepath):
    checkpoint = torch.load(filepath)
    learning_curve = checkpoint['learning_curve']
    source_accuracies, target_accuracies, classification_losses, discriminator_losses, encoder_losses = zip(*learning_curve)

    epochs = range(1, len(learning_curve) + 1)

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.plot(epochs, source_accuracies, label='Source')
    plt.plot(epochs, target_accuracies, label='Target')
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.title("Accuracies")

    plt.subplot(2, 2, 2)
    plt.plot(epochs, classification_losses, label='Classification Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.title("Classification Loss")

    plt.subplot(2, 2, 3)
    plt.plot(epochs, discriminator_losses, label='Discriminator Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.title("Discriminator Loss")

    plt.subplot(2, 2, 4)
    plt.plot(epochs, encoder_losses, label='Encoder Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.title("Encoder Loss")

    plt.tight_layout()
    plt.show()
```

```python
# Plot learning curves for the best source and target models
plot_learning_curve("learning_curve.pth")
#plot_learning_curve("best_target_model.pth")
```
